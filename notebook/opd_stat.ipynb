```python
#健保平均門診就診率: 健保平均每十萬人口門診就診率(門診患者人數 / 年中人口數* 100,000)
import pandas as pd
from typing import Union, Tuple, Optional
import matplotlib.pyplot as plt
from pathlib import Path

from matplotlib.axes import Axes

plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] 
plt.rcParams['axes.unicode_minus'] = False   

import xml.etree.ElementTree as XET
tree = XET.parse('/Users/wei/visit_statistics/test_file/健保平均門診就診率-按性別、年齡、疾病別(第10版)分.XML')   #以XET套件載入XML檔案
root = tree.getroot()         #取得XML表格 

df_cols=["Gender","Age","Diseases"]
rows=[]
for i in range(len(root)): 
    
    x=root[i][0].text, root[i][1].text, root[i][2]
    Gender=root[i][0].text
    Age=root[i][1].text
    Diseases=root[i][2].text
    
    
    rows.append({"Gender":Gender,"Age":Age,"Diseases":Diseases})

df=pd.DataFrame(rows,columns=df_cols)
```

```python
def parse_xml(df: pd.DataFrame, 
              gender: str,
              age_range: Optional[Tuple[int, int]] = None) -> pd.DataFrame:
 

    df.columns = ["Gender", "Age", "Diseases",]
    df["Gender"].replace({"女": "F", "男": "M"}, inplace=True)
   

    if age_range is None:  
        m = df['Age'] == "總計"
    elif age_range[1]-age_range[0] != 4:
        print ("please do the age difference to equal four.")
    elif age_range[0] %5 !=0:     
        print ("please notice that the first age is a multiple of 5.")
    else:
        if age_range[0] != 85:    
            m = df['Age'] == f'{age_range[0]}~{age_range[1]}歲'   #
        else:  # above 85 years-old
            m = df['Age'] == f'{age_range[0]}歲以上'
    df = df[m]

    gm = df['Gender'] == gender
    df = df[gm]
 
    return df
```
